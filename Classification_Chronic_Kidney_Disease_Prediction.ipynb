```python
import pandas as pd
```

```python
dataset = pd.read_csv('CKD.csv')
dataset.describe()
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset.head()
```

```python
dataset = dataset.astype(int)
dataset.head()
```

```python
independent = dataset.drop(['classification_yes'], axis=1)
dependent = dataset['classification_yes']
```

```python
independent.head()
```

```python
dependent.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent,dependent, test_size = 0.3, random_state = 0, stratify=dependent)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
param_DTC = {'criterion' : ['gini', 'entropy', 'log_loss'],
              'splitter' : ['best', 'random'],
               'max_features' : ['sqrt', 'log2']}

Model_DTC = GridSearchCV(DecisionTreeClassifier(), param_DTC, refit=True, verbose=3, n_jobs=-1, error_score='raise')
Model_DTC = Model_DTC.fit(x_train,y_train)
```

```python
best_model_DTC = Model_DTC.best_params_
print(best_model_DTC)
```

```python
y_pred_DTC = Model_DTC.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_DTC = confusion_matrix(y_test, y_pred_DTC)
print(cm_DTC)
```

```python
from sklearn.metrics import classification_report
clf_report_DTC = classification_report(y_test, y_pred_DTC)
print(clf_report_DTC)
```

```python
from sklearn.ensemble import RandomForestClassifier
param_RFC = {'n_estimators' : [10,50,100], 'criterion' : ['gini', 'entropy', 'log_loss'], 'max_features' : ['sqrt', 'log2']}

Model_RFC = GridSearchCV(RandomForestClassifier(), param_RFC, refit=True, verbose=3, n_jobs=-1, error_score='raise')
Model_RFC = Model_RFC.fit(x_train,y_train)
```

```python
best_model_RFC = Model_RFC.best_params_
print(best_model_RFC)
```

```python
y_pred_RFC = Model_RFC.predict(x_test)
```

```python
cm_RFC = confusion_matrix(y_test, y_pred_RFC)
print(cm_RFC)
```

```python
clf_report_RFC = classification_report(y_test, y_pred_RFC)
print(clf_report_RFC)
```
